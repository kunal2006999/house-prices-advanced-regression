```python
import pandas as pd
import numpy as np

# Load the data
train = pd.read_csv("C:/Users/kunal/House Prices - Advanced Regression Techniques/data/train.csv")
test = pd.read_csv("C:/Users/kunal/House Prices - Advanced Regression Techniques/data/test.csv")

# Show first few rows
print(train.shape)
train.head()
```

```python
from sklearn.metrics import mean_squared_error
def rmse_log(y_true, y_pred):
    return np.sqrt(mean_squared_error(np.log1p(y_true), np.log1p(y_pred)))
```

```python
y = train["SalePrice"]
X = train.drop(columns=["SalePrice", "Id"])
test_ids = test["Id"]
X_test = test.drop(columns=["Id"])
```

```python
X = train.drop(columns=['Id', 'SalePrice'])
y = np.log1p(train['SalePrice'])
```

```python
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
ord_cols = [
    'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
    'HeatingQC', 'KitchenQual', 'FireplaceQu',
    'GarageQual', 'GarageCond', 'BsmtExposure',
    'BsmtFinType1', 'BsmtFinType2', 'Functional',
    'GarageFinish', 'PavedDrive', 'Fence'
]
exset = set(categorical_features) - set(ord_cols)
cat_cols = list(exset)
```

```python
cat_cols_fill = ["PoolQC", "Alley", "Fence", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "MasVnrType"]
for col in cat_cols_fill:
    X[col] = X[col].fillna("None")
num_cols_fill = ["MasVnrArea", "GarageYrBlt", "BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF", "BsmtFullBath", "BsmtHalfBath", "GarageCars", "PoolArea"]
for col in num_cols_fill:
    X[col] = X[col].fillna(0)
X["LotFrontage"] = X.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

Special-case imputations

Certain categorical features where NA means absence of a feature (e.g., no pool, no garage, no basement) were filled with "None".

Numeric features where NA implies zero area or count were filled with 0.

LotFrontage was imputed using the median frontage within each Neighborhood, preserving local patterns.

```python
from sklearn.impute import SimpleImputer
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')
ord_imputer = SimpleImputer(strategy='constant', fill_value='None')

ord_categories = [
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # ExterQual
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # ExterCond
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # BsmtQual
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # BsmtCond
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # HeatingQC
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # KitchenQual
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # FireplaceQu
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # GarageQual
    ['None', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],   # GarageCond
    ['None', 'No', 'Mn', 'Av', 'Gd'],         # BsmtExposure
    ['None', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'], # BsmtFinType1
    ['None', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'], # BsmtFinType2
    ['None', 'Sal', 'Sev', 'Maj2', 'Maj1', 'Mod', 'Min2', 'Min1', 'Typ'], # Functional
    ['None', 'Unf', 'RFn', 'Fin'],            # GarageFinish
    ['None', 'N', 'P', 'Y'],                  # PavedDrive
    ['None', 'MnWw', 'GdWo', 'MnPrv', 'GdPrv'] # Fence
]

from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
cat_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
ord_encoder = OrdinalEncoder(categories=ord_categories)


from sklearn.pipeline import Pipeline

num_pipe = Pipeline([
    ('imputer', num_imputer),
    ('scaler', StandardScaler())
])

ord_pipe = Pipeline([
    ('imputer', ord_imputer),
    ('encoder', ord_encoder)
])

cat_pipe = Pipeline([
    ('imputer', cat_imputer),
    ('encoder', cat_encoder)
])

from sklearn.compose import ColumnTransformer

preproc = ColumnTransformer(
    transformers=[
        ('num', num_pipe, numeric_features),
        ('ord', ord_pipe, ord_cols),
        ('cat', cat_pipe, cat_cols)
    ],
    remainder='passthrough'
)

```

```python

from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline

model_pipe = Pipeline([
    ('preproc', preproc),
    ('model', XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    ))
])
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

SEED = 42
np.random.seed(SEED)

kf = KFold(n_splits=5, shuffle=True, random_state=SEED)
oof_preds = np.zeros(len(X))
cv_scores = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
    print(f"Fold {fold+1}")
    
    # Split train/validation data
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Fit model on training fold
    model_pipe.fit(X_train, y_train)
    
    # Predict on validation fold
    val_preds = model_pipe.predict(X_val)
    oof_preds[val_idx] = val_preds
    
    # Evaluate RMSE
    rmse = np.sqrt(mean_squared_error(y_val, val_preds))
    cv_scores.append(rmse)
    print(f"Fold {fold+1} RMSE: {rmse:.4f}")
```

```python
mean_rmse = np.mean(cv_scores)
std_rmse = np.std(cv_scores)
print(f"\nCV RMSE: {mean_rmse:.4f} ± {std_rmse:.4f}")
```

```python
index = np.arange(50)

plt.figure(figsize=(10, 6))

plt.plot(index, y.tolist()[:50], color='black', linestyle='--', label='Actual Prices (Reference)', linewidth=1.5)

plt.plot(index, oof_preds[:50], color='blue', label='Model A Predictions', linewidth=2)

plt.title('Comparison of Two Model Predictions vs. Actual Prices')
plt.xlabel('Observation Index')
plt.ylabel('Price')
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

rf_pipe = Pipeline([
    ('preproc', preproc),
    ('model', RandomForestRegressor(random_state=42))
])

param_grid = {
    'model__n_estimators': [100, 200, 500],
    'model__max_depth': [5, 10, None],
    'model__min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    estimator=rf_pipe,
    param_grid=param_grid,
    cv=5,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)
grid_search.fit(X, y)
print(grid_search.best_params_)
print(-grid_search.best_score_)
```

```python
xgb_pipe = Pipeline([
    ('preproc', preproc),
    ('model', XGBRegressor(random_state=42))
])

xgb_param_grid = {
    'model__n_estimators': [100, 300, 500],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__max_depth': [3, 4, 5],
    'model__subsample': [0.8, 0.9],
    'model__colsample_bytree': [0.8, 0.9]
}

xgb_grid_search = GridSearchCV(
    estimator=xgb_pipe,
    param_grid=xgb_param_grid,
    cv=5,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1
)
xgb_grid_search.fit(X, y)
print(xgb_grid_search.best_params_)
print(-xgb_grid_search.best_score_)
```

```python
best_xgb = xgb_grid_search.best_estimator_
best_xgb.fit(X, y)
```

```python
X_test = test.copy()

y_pred = best_xgb.predict(X_test)
y_pred_final = np.expm1(y_pred)
```

```python
submission = pd.DataFrame({
    "Id": test["Id"],
    "SalePrice": y_pred_final
})
submission.to_csv("submission.csv", index=False)
print("✅ submission.csv is ready!")
```
